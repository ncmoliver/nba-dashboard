```python
def get_team_players(data, year, team):
    """
    Filters the dataset by year and team and returns the players on the team.

    Args:
        data (pd.DataFrame): The dataset containing player, team, and season information.
        year (int): The year (season) to filter by.
        team (str): The team name to filter by.

    Returns:
        list: A list of players who were on the specified team during the given year.
    """
    # Filter the dataset by the selected year and team
    filtered_data = data[(data['year'] == year) & (data['Team'] == team)]
    
    # Extract the list of players
    players = filtered_data['PLAYER'].unique().tolist()
    
    return players
```

```python
def get_data_by_year_range(data, start_year, end_year):
    """
    Filters the dataset to include only rows within the specified year range.

    Args:
        data (pd.DataFrame): The dataset containing season information.
        start_year (int): The start year of the range (inclusive).
        end_year (int): The end year of the range (inclusive).

    Returns:
        pd.DataFrame: A filtered DataFrame containing data from the specified year range.
    """
    # Filter the dataset for the specified year range
    filtered_data = data[(data['year'] >= start_year) & (data['year'] <= end_year)]
    return filtered_data
```

```python
import pandas as pd

def get_condensed_player_stats(data, team, year):
    """
    Returns a DataFrame of players on a team for a specific year,
    with each player's stats averaged if there are duplicates.

    Args:
        data (pd.DataFrame): The dataset containing player, team, and season information.
        team (str): The selected team.
        year (int): The selected year.

    Returns:
        pd.DataFrame: A DataFrame containing the averaged stats for each player.
    """
    # Validate inputs
    if not team:
        raise ValueError("Please provide a valid team.")

    if not year:
        raise ValueError("Please provide a valid year.")

    # Filter the dataset for the provided team and year
    team_players = data[(data['Team'] == team) & (data['year'] == year)]

    # Check if there are players for the provided team and year
    if team_players.empty:
        raise ValueError(f"No players found for the team: {team} in the year: {year}")

    # Group by PLAYER and calculate the mean for numeric columns
    numeric_columns = team_players.select_dtypes(include='number').columns
    condensed_players = team_players.groupby(['PLAYER', 'Pos_x'], as_index=False)[numeric_columns].mean()

    return condensed_players
```
